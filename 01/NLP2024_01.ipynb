## Bevezetés

## Mi az NLP?

Az NLP a nyelvészet és a gépi tanulás egy területe, amely az emberi nyelvre  összpontosít. Az NLP feladatok célja nem csak az egyes szavak megértése, **hanem a szavak környezetének, kontextusának megértése is**.

## Gyakori NLP feladatok

Íme néhány példa a tipikus NLP feladatokra:

* **Teljes mondatok osztályozása:**
    * Véleményelemzés egy szövegből
    * Spam e-mailek azonosítása
    * Mondatok nyelvtani helyességének ellenőrzése
    * Két mondat logikai összefüggésének meghatározása
* **Egyes szavak osztályozása egy mondatban:**
    * A mondat grammatikai összetevőinek azonosítása (főnév, ige, melléknév)
    * Névnevek (személy, hely, szervezet) azonosítása
* **Szöveg tartalom generálása:**
    * Befejezetlen kérdések automatikus kiegészítése
    * Szövegrészek kitöltése elrejtett szavakkal
* **Válasz kinyerése egy szövegből:**
    * Válaszadás egy adott kérdésre egy megadott szöveg alapján
* **Új mondat generálása egy bemeneti szövegből:**
    * Szöveg fordítása egy másik nyelvre
    * Szöveg összefoglalása

Az NLP nem korlátozódik az írott szövegre. Ez magában foglalja a beszédfelismerést és a számítógépes látást is, például hangminták átírását vagy képek leírását.

##Pipeline-okkal dolgozni

A  Transformers könyvtár egyik eleme a `pipeline()` függvény. Ez egy modellt kapcsol össze a szükséges előfeldolgozási és utófeldolgozási lépésekkel, lehetővé téve számunkra, hogy bármilyen szöveget közvetlenül beírjunk és érthető választ kapjunk:


````python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for this course my whole life.")
````

````python
classifier(
    ["I've been waiting for this course my whole life.", "I hate this so much!"]
)
````

 **Alapértelmezés szerint ez a pipeline egy olyan előre betanított modellt választ ki, amelyet angol nyelvű véleményelemzésre finomhangoltak. A modellt a classifier objektum létrehozásakor tölti le és tárolja a rendszer a gyorsítótárban. Ha újra futtatja a parancsot, a gyorsítótárban tárolt modellt fogja használni, és nem kell újra letöltenie a modellt.**

**Három fő lépésből álló folyamat zajlik le, amikor szöveget küld a pipeline-nak:**

1. **A szöveg előfeldolgozásra kerül egy olyan formátumba, amelyet a modell képes értelmezni.**
2. **A feldolgozott bemeneteket továbbítja a modellnek.**
3. **A modell előre jelzései utófeldolgozásra kerülnek, hogy értelmezhesd őket.**

**Néhány a jelenleg elérhető pipeline-ok közül:**

* feature-extraction (a szöveg vektoros reprezentációjának lekérése)
* fill-mask (maszkolt szavak kitöltése)
* ner (név entitások felismerése)
* question-answering (kérdések megválaszolása)
* sentiment-analysis (véleményelemzés)
* summarization (összefoglalás)
* text-generation (szöveg generálása)
* translation (fordítás)
* zero-shot-classification (zero-shot osztályozás)

**Nézzük meg néhányat ezek közül!**

**Zero-shot osztályozás**

Egy nagyobb kihívást jelentő feladattal kezdjük, ahol olyan szövegeket kell osztályozni, amelyek nincsenek felcímkézve. Ez egy gyakori forgatókönyv a valós projektekben, mivel a szövegek felcímkézése általában időigényes és szakértelmet igényel. Erre a felhasználási esetre a zero-shot-classification pipeline igen hatékony: lehetővé teszi a kategorizáláshoz használni kívánt címkék megadását, így nem kell a előre betanított modell címkéire támaszkodnia. Láttad már, hogyan tudja a modell egy mondatot pozitív vagy negatív kategóriába sorolni e két címke segítségével - de képes a szöveg kategorizálására is, tetszőleges más címkekészlet használatával.


````python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
````

A pipeline-t zero-shot-nak nevezik, mert a használatához nem kell finomhangolnia a modellt a saját adatain. Bármilyen kívánt címkelistára közvetlenül vissza tud adni valószínűségi pontszámokat!

Szöveg generálás

Most nézzük meg, hogyan használható egy pipeline szöveg generálására. A lényeg, hogy megadsz egy kezdő mondatot, és a modell automatikusan kiegészíti azt a hátralévő szöveg létrehozásával. Névfelismerés

A névfelismerés (NER) egy olyan feladat, ahol a modellnek meg kell találnia a beviteli szöveg azon részeit, amelyek entitásokra utalnak, mint például személyek, helyek vagy szervezetek. Nézzünk egy példát:


````python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
````

Itt a modell helyesen azonosította, hogy Sylvain egy személy (PER), a Hugging Face egy szervezet (ORG), a Brooklyn pedig egy helyszín (LOC).

A `grouped_entities=True` opciót adjuk át a pipeline létrehozó függvénynek, hogy megmondjuk a pipeline-nak, hogy csoportosítsa újra a mondat azon részeit, amelyek ugyanahhoz az entitáshoz tartoznak: itt a modell helyesen csoportosította a "Hugging" és "Face" szavakat egyetlen szervezetként, annak ellenére, hogy a név több szóból áll. Valójában, ahogyan a következő fejezetben látni fogjuk, az előfeldolgozás még egyes szavakat is kisebb részekre bont. Például a "Sylvain" négy részre van osztva: "S", "##yl", "##va" és "##in". Az utófeldolgozási lépésben a pipeline sikeresen újracsoportosította ezeket a részeket.

✏️ Próbáld ki! Keress a Model Hubon egy olyan modellt, amely képes angol nyelvű beszédrészelemzésre (általában POS rövidítéssel jelölve). Mit jósol ez a modell a fenti példamondatra?

````python
import spacy

filename = "piszkosfred_reszlet.txt"
file = open(filename, "r", encoding="utf-8")
text = file.read()
text = text.replace("\n", " ")
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
````

````python
words = [token.text for token in doc]
pos = [token.pos_ for token in doc]
word_pos_tuples = list(zip(words, pos))
````
