```python
# Importált modulok
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import time  # Új modul importálása a késleltetéshez
```

```python
# Táblák definiálása
web_pages = pd.DataFrame(columns=['url', 'title'])
articles = pd.DataFrame(columns=['page_url', 'url', 'title', 'text'])
```

```python
# Adatok gyűjtése és mentése
urls = ['https://index.hu/', 'https://origo.hu/', 'https://www.portfolio.hu/', 'https://www.napi.hu/', 'https://444.hu/', 'https://24.hu/', 'https://telex.hu/', 'https://mno.hu/', 'https://magyarnemzet.hu/', 'https://444.hu/']
keyword = 'tűz'

for url in urls:
    try:
        # Kérés küldése az URL-re
        r = requests.get(url)
        # Hiba ellenőrzése
        r.raise_for_status()

        # BeautifulSoup objektum létrehozása a HTML-ből
        soup = BeautifulSoup(r.content, 'html.parser')

        # A weboldal címe
        title = soup.find('title').text.strip()

        # Oldal hozzáadása a web_pages táblához
        web_pages = web_pages.append({'url': url, 'title': title}, ignore_index=True)

        # Cikkek keresése
        for link in soup.find_all('a'):
            # A cikk URL-je
            article_url = link.get('href')

            # Ellenőrzi, hogy a kulcsszó szerepel-e a link szövegében
            if article_url is not None and keyword in link.text.lower():

                # Ciklus az articles táblához
                try:
                    # Kérés küldése a cikk URL-jére
                    article_r = requests.get(article_url)
                    # Hiba ellenőrzése
                    article_r.raise_for_status()

                    # BeautifulSoup objektum létrehozása a cikk HTML-jéből
                    article_soup = BeautifulSoup(article_r.content, 'html.parser')

                    # A cikk szövege
                    article_text = article_soup.get_text(separator='\n')

                    # Cikkek hozzáadása az articles táblához
                    articles = articles.append({'page_url': url, 'url': article_url, 'title': link.text.strip(), 'text': article_text}, ignore_index=True)
                except requests.exceptions.RequestException as e:
                    # Hibaüzenet kiírása
                    print(f"Hiba a cikk letöltése közben: {e}")
                    # Tovább a következő cikkre
                    continue

    except requests.exceptions.RequestException as e:
        # Hibaüzenet kiírása
        print(f"Hiba a weboldal letöltése közben: {e}")
        # Tovább a következő weboldalra
        continue

    # Késleltetés az újabb kérések között
    time.sleep(1)  # 1 másodperces késleltetés
```
